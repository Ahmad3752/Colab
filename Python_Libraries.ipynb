```python
import matplotlib.pyplot as plt
import numpy as np
# Sample data
x = np.linspace(0, 10, 100)
y = np.cos(x)
# Create plot
plt.plot(x, y, label="Cos Wave")
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 6))
x = np.linspace(0, 10, 100)
ax[0].plot(x, np.sin(x), 'r', label="Sine Wave")
ax[1].plot(x, np.cos(x), 'b', label="Cosine Wave")
ax[0].set_title("Sine Function")
ax[1].set_title("Cosine Function")
for a in ax:
    a.set_xlabel("X-axis")
    a.set_ylabel("Y-axis")
    a.legend()
    a.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/datasets/s-and-p-500/master/data/data.csv"
df = pd.read_csv(url)

# Convert 'Date' to datetime and set as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Check column names
print(df.columns)  # Verify available columns

# Plot stock prices
plt.figure(figsize=(10,5))
df['50-day MA'] = df['SP500'].rolling(window=50).mean()
plt.plot(df.index, df['SP500'], color='b', linestyle='-', label="S&P 500")  # Ensure 'Close' exists
plt.plot(df.index, df['50-day MA'], color='r', linestyle='--', label="50-day Moving Average")

plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("S&P 500 Closing Prices Over Time")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.cos(x)

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

data = np.random.randn(1000)
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

# Line plot
ax[0, 0].plot(x, np.sin(x), 'r')
ax[0, 0].set_title("Sine Wave")

# Bar chart
ax[0, 1].bar(categories, values, color='b')
ax[0, 1].set_title("Bar Chart")

# Scatter plot
ax[1, 0].scatter(x, y, c='g')
ax[1, 0].set_title("Scatter Plot")

# Histogram
ax[1, 1].hist(data, bins=20, color='purple', edgecolor='black')
ax[1, 1].set_title("Histogram")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips = sns.load_dataset("tips")
print(tips.head(5))
print(tips.columns)
print(tips.dtypes)
print(tips.isna().sum())
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker')
plt.title("Total Bill vs Tip")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.histplot(tips['total_bill'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Distribution of Total Bill")

# Boxplot: Compare total_bill across different days
sns.boxplot(x='day', y='total_bill', data=tips, ax=axes[0, 1])
axes[0, 1].set_title("Total Bill Across Different Days")

# Scatterplot: Relationship between total_bill and tip
sns.scatterplot(x='total_bill', y='tip', data=tips, ax=axes[1, 0])
axes[1, 0].set_title("Total Bill vs. Tip")

# Barplot: Average tip amount for each day
sns.barplot(x='day', y='tip', data=tips, ax=axes[1, 1])
axes[1, 1].set_title("Average Tip Amount by Day")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Task 1: Heatmap showing correlations between numerical features
numerical_cols = tips.select_dtypes(include=['float64', 'int64']).columns
corr = tips[numerical_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5, ax=axes[0, 0])
axes[0, 0].set_title("Correlation Heatmap")

# Task 2: Regression plot to analyze relationship between total_bill and tip
sns.regplot(x='total_bill', y='tip', data=tips, ax=axes[0, 1])
axes[0, 1].set_title("Regression Plot: Total Bill vs. Tip")

# Task 3: FacetGrid for scatter plots based on the 'sex' column
g = sns.FacetGrid(tips, col="sex", height=5)
g.map_dataframe(sns.scatterplot, x='total_bill', y='tip')
g.set_axis_labels('Total Bill', 'Tip')
g.set_titles("{col_name} Scatter Plot")
plt.tight_layout()

# Show all the plots
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Task 1: Add a title and labels to a scatterplot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()

# Task 2: Use a different color palette for a barplot
sns.barplot(x='day', y='total_bill', data=tips, palette='pastel')
plt.title("Average Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill ($)")
plt.show()

# Task 3: Modify the legend of a scatterplot
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
plt.title("Total Bill vs Tip by Gender")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.legend(title="Gender")
plt.show()
```
