```python
df1=pd.read_csv('/content/Task_2_Feedback_Data.csv')
weights = {'Excellent': 5, 'Very Good': 4, 'Good': 3, 'Average': 2, 'Poor': 1}
df1.head()
```

```python
def Score(row):
  return (
    row['Excellent'] * 5 +
    row['Very Good'] * 4 +
    row['Good'] * 3 +
    row['Average'] * 2 +
    row['Poor'] * 1
)
def weightedSum(row):
  sum=(
    row['Excellent'] +
    row['Very Good']  +
    row['Good']  +
    row['Average']  +
    row['Poor']
)
  return row['SatisfactionScore']/sum

df1['SatisfactionScore']=df1.apply(Score,axis=1)
df1['WeightedScore']=df1.apply(weightedSum,axis=1)
df1.head()
```

```python
me=df1['WeightedScore'].mean()
```

```python
sns.barplot(data=df1,x='Unnamed: 0',y='WeightedScore')
plt.xticks(rotation=90)
plt.xlabel('Questions')
plt.axhline(me, color='green', linestyle='dashed', label=f'Mean: {me:.2f}')
plt.ylabel('Weighted Score')
plt.title('Weighted Scores by Question')
plt.legend();
```

```python
df2 = pd.read_csv('/content/Task_3_Accident_Data.csv')
df2.head()
```

```python
plt.figure(figsize=(12, 8))

sns.scatterplot(x=df2['Day'], y=df2['Place A'], label='Place A', color='blue', s=100, marker='o')
sns.scatterplot(x=df2['Day'], y=df2['Place B'], label='Place B', color='green', s=100, marker='s')
sns.scatterplot(x=df2['Day'], y=df2['Place C'], label='Place C', color='red', s=100, marker='^')

plt.title('Accident Distribution Over Time')
plt.xlabel('Day')
plt.ylabel('Number of Accidents')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)

plt.show()

skewness_a = skew(df2['Place A'])
skewness_b = skew(df2['Place B'])
skewness_c = skew(df2['Place C'])

print(f"Skewness for Place A: {skewness_a:.2f}")
print(f"Skewness for Place B: {skewness_b:.2f}")
print(f"Skewness for Place C: {skewness_c:.2f}")

total_accidents = {
    'Place A': df2['Place A'].sum(),
    'Place B': df2['Place B'].sum(),
    'Place C': df2['Place C'].sum()
}

# Identify highest and lowest accident risk locations
most_risky_place = max(total_accidents, key=total_accidents.get)
least_risky_place = min(total_accidents, key=total_accidents.get)

print(f"\nHighest accident risk: {most_risky_place} with {total_accidents[most_risky_place]} accidents.")
print(f"Lowest accident risk: {least_risky_place} with {total_accidents[least_risky_place]} accidents.")
```

```python
df4=pd.read_csv('/content/Task_5_Salaries_Data.csv')
df4.head()
```

```python
from scipy.stats import skew

salary_skewness = skew(df4['Salaries (Thousands)'])
print(f"Skewness of Salary Distribution: {salary_skewness}")
```

```python
import numpy as np
df4_sorted = df4.sort_values(by="Salaries (Thousands)")

df4_sorted['Cumulative Frequency'] = np.arange(1, len(df4_sorted) + 1)

plt.figure(figsize=(10, 6))

# Use plot instead of bar for ogive
plt.plot(df4_sorted["Salaries (Thousands)"],
         df4_sorted["Cumulative Frequency"],
         marker='o', linestyle='-', color='blue', markersize=5)

plt.title("Ogive Curve of Salaries")
plt.xlabel("Salaries (Thousands)")
plt.ylabel("Cumulative Frequency")
plt.grid(True, linestyle="--", alpha=0.7)

plt.xticks(df4_sorted["Salaries (Thousands)"], rotation=45)  # Rotate x-axis labels if needed
plt.tight_layout()  # Adjust layout to prevent clipping

plt.show()
```
