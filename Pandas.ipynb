```python
import numpy as np
import pandas as pd

people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'],
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'],
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, np.nan, np.nan],
    'age': ['33', '55', '63', '36', None, 'Missing', np.nan]
}

df2 = pd.DataFrame(people)
df2
```

```python
#dealing with missing values INDEX;ROW
#df2.dropna(axis='index',how='any',inplace=True) # by default how =any but the problem is that if any value is missing it wil ldelete so i will change how to all
#df2.dropna(axis='index',how='all',inplace=True) # now row with all misswing value will be drop[]
#df2

# acha kabhi kabhi misaal k tor par hume email ko sabse important rakhna toh us column mai agr missing values thab drop karo matlab subsetting
df2.dropna(axis='index',how='all',subset=['email'],inplace=True)
df2
```

```python
#filling
df2.dtypes
#changing age data type pehle hum missing ko np.nan mai karenge q k o float hai
df2['age']=df2['age'].replace({'NA':np.nan,'MISSING':np.nan},inplace=True)
df2['age']=df2['age'].astype(float) #acha agar int mai karenge toh error 1 k np.nan float hai int pe error
```
